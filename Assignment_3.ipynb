```python
import pandas as pd
import requests
```

```python
url = "https://api.tvmaze.com/shows/1/episodes"

payload = {}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)

data = response.json()
type(data)
```

```python
df = pd.DataFrame(data)
df.head(10)
```

```python
df['rating'] = df['rating'].apply(lambda x: x['average'])
df['image'] = df['image'].apply(lambda x: x['medium'])
df['_links'] = df['_links'].apply(lambda x: x['self']. get('href'))
df['summary'] = df['summary'].apply(lambda x: x.replace('<p>','').replace('</p>',''))
```

```python
filterRating = df[df['rating'] > 5]
group = df.groupby(['season']).agg(season_count=('season', 'count')).reset_index()
```
